```python
# Initilise empty matrix
n = len(location_df)
pairwise_dist_matrix = np.zeros((n, n))  # Initializes with zeros
```

```python
len(distances_list)
```

```python
distances_list_binned = []
lower_bound = 0
upper_bound = 0

for i in range(31, 0, -1):
    upper_bound += i
    print(lower_bound, upper_bound, upper_bound-lower_bound)
    distances_list_binned.append(distances_list[lower_bound:upper_bound])
    lower_bound += i
```

```python
for i in range(n):
    print(f'Populating distance from {location_df['location_name'][i]} to elsewhere {(i,n)}')
    for j in range(i, n):
        # print(i,j)
        pairwise_dist_matrix[i][j] = distances_list_binned[i][j-i]
        pairwise_dist_matrix[j][i] = distances_list_binned[i][j-i]
```

```python
fp = 'data/pairwise_dist_matrix.pkl'
helpers.write_pickle(fp, pairwise_dist_matrix)
```
